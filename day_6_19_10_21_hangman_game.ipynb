```python
import random # import to import words and to chose random words from word list
def hangman(): #function called and declared 
  list_of_words=['welcome', 'shivmurat', 'nandesh', 'india', 'brillica', 'last'] #word list
  word =random.choice(list_of_words) #random.choice is used to choice random words from list
  turns=8 #attempts
  guessmade='' #guess is blank
  valid_entry=set('abcdefghijklmnopqrstuvwxyz') # input characters to input from this set 
  # it can e also be an Array
  #set is used to write array is "valid_entry=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']"

  while len(word)>0: # mearuring the word list [ using len(word) ] 
    main_word=""   

    for letter in word: #using for loop and condition is to get letters from word list
      if letter in guessmade: #using if and condition to get blank from ( guessmade='' )
        main_word=main_word+letter #word + letters = blank space
      else:
        main_word=main_word+"_ " # __ will be displayed 

    if main_word == word: # condition when we won the game
      print(main_word)
      print("\n!!!!! Congratulations You Won !!!!!\n")
      break # to stop the loop or if condition / to stop the program
#########################################################
    print("Guess the words", main_word)  # taking input 
    guess = input()

    if guess in valid_entry: # taking characters from [ valid_entry ]
      guessmade = guessmade+guess # making guesses 
    else:
      print("Enter Valid character")# if input is apart from Alphabet character
      guess=input()
########################################################
    if guess not in word:
      turns=turns-1 #decrement after wrong guess
      
      # This is hangman after makig wrong guesses # 
      if turns==7: #1st wrong guess
          print("7 turns left!!! This is 1st mistake") 
          print("-------------------------------")
      if turns==6: #1st wrong guess
          print("6 turns left!!!")
          print("-------------------------------")
          print("              \O/              ")
          print("               |               ")
          print("              / \              ")
      if turns==5: #2nd wrong guess
          print("5 turns left!!!")
          print("-------------------------------")
          print("         |                     ")
          print("         |                     ")
          print("         |     \O/             ")
          print("         |      |              ")
          print("         |     / \             ")
      if turns==4: #3rd wrong guess
          print("4 turns left!!!")
          print("-------------------------------")
          print("         _____________         ")
          print("         |                     ")
          print("         |                     ")
          print("         |     \O/             ")
          print("         |      |              ")
          print("         |     / \             ")
      if turns==3: #4th wrong guess
          print("3 turns left!!!")
          print("-------------------------------")
          print("         _____________         ")
          print("         | /                   ")
          print("         |/                    ")
          print("         |     \O/             ")
          print("         |      |              ")
          print("         |     / \             ")
      if turns==2: #5th wrong guess
          print("2 turns left!!! Hurry up")
          print("-------------------------------")
          print("         _____________         ")
          print("         | /    |              ")
          print("         |/                    ")
          print("         |     \O/             ")
          print("         |      |              ")
          print("         |     / \             ")
      if turns==1: #7th wrong guess
          print("Only 1 turns left!!! You are on your last breath")
          print("-------------------------------")
          print("         _____________         ")
          print("         | /    |              ")
          print("         |/     |              ")
          print("         |     \O/             ")
          print("         |      |              ")
          print("         |     / \             ")
      if turns==0: #8th wrong guess
          print("Only 1 turns left!!! You are on your last breath")
          print("-------------------------------")
          print("         _____________         ")
          print("         | /    |              ")
          print("         |/     |              ")
          print("         |      O              ")
          print("         |     /|\             ")
          print("         |     / \             ")
          print("\n!!!!! You loose !!!!!  -> **Looooser**")
          print("You let a Hangman Die Looser")
          break # to stop the loop or if condition / to stop the program
########################################################
#this is the main program to run 1st
name = input("Enter Your Name ->\t") #player name
print("Welcome",name,"!!!") #welcoming player
print("------------------------------------------------------")
print("Try to guess the word in 8 attempts") #warning for attemps to be used
hangman()      #function calling

```
